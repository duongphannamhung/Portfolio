# B. explore data đơn hàng

## 1. Xuất dữ liệu từ source ERP để phân tích

```python

data_df=data.drop(columns=['tenant_id','session_id','identity.keys','identity.vals','tags.keys','tags.vals','scope']).reset_index()
erp_df1=data_df[(data_df['event_name'].str.contains('erp_item_order_completed'))|(data_df['event_name'].str.contains('erp_order_completed'))]
```

```python
tesst=erp_df1[['anonymous_id','event_name']]
_count=tesst.groupby('anonymous_id').agg({"event_name": "nunique"}).reset_index()
_count
```

```python
_count_df=_count[_count['event_name']==2]
list_cus=_count_df['anonymous_id'].unique().tolist()
```

```python
erp_df=erp_df1[erp_df1['anonymous_id'].isin(list_cus)]
erp_df.groupby(['anonymous_id','event_name']).count()
```

### 2.1. Tách các trường chứa chuỗi thành các dòng

### Chuyển dữ liệu dạng text thành dạng list []

```python
def literal_return(val):
    try:
        return ast.literal_eval(val)
    except ValueError:
        return (val)
erp_df['str_properties.vals']=erp_df['str_properties.vals'].apply(literal_return)
erp_df['str_properties.keys']=erp_df['str_properties.keys'].apply(literal_return)
erp_df['num_properties.keys']=erp_df['num_properties.keys'].apply(literal_return)
erp_df['num_properties.vals']=erp_df['num_properties.vals'].apply(literal_return)
```

### 2.2 Chia dữ liệu thành 2 nhóm: 
- Dạng string chứa: str_properties.keys, str_properties.vals
- Dạng numeric chứa: num_properties.keys, num_properties.vals

```python
data_str=erp_df[['id','anonymous_id','event_name','str_properties.keys','str_properties.vals','at']]
# data_str=data_str.drop_duplicates()
data_str_df=data_str.set_index(['id','anonymous_id','event_name','at']).apply(lambda x: x.apply(pd.Series).stack()).reset_index()
```

```python
data_str_df=data_str_df.drop(columns='level_4').rename(columns={'str_properties.keys':'keys','str_properties.vals':'vals'})
```

```python
data_str_df.head()
```

```python
data_str_df=data_str_df.rename(columns={'str_properties.keys':'keys','str_properties.vals':'vals'})

data_str_df.groupby('anonymous_id').count()
```

```python
data_num=erp_df[['id','anonymous_id','event_name','num_properties.keys','num_properties.vals','at']]
data_num_df=data_num.set_index(['id','anonymous_id','event_name','at']).apply(lambda x: x.apply(pd.Series).stack()).reset_index()
data_num_df=data_num_df.drop(columns='level_4').rename(columns={'num_properties.keys':'keys','num_properties.vals':'vals'})
data_num_df.head(2)
```

```python
data_num_df.groupby('anonymous_id').count()
```

## 2. Phân tích các feature

```python
# số lượng order của mỗi khách
erp_order_str=data_str_df[data_str_df['event_name'].isin(['erp_order_completed'])]
erp_order_str_pivot=erp_order_str.pivot(index=['id','anonymous_id','event_name','at'],columns='keys',values='vals').reset_index()
erp_order_str_pivot=erp_order_str_pivot.replace(np.nan,0)
erp_order_str_pivot
```

```python
erp_order_num=data_num_df[data_num_df['event_name'].isin(['erp_order_completed'])]
erp_order_num_pivot=erp_order_num.pivot(index=['id','anonymous_id','event_name','at'],columns='keys',values='vals').reset_index()
erp_order_num_pivot=erp_order_num_pivot.replace(np.nan,0)
erp_order_num_pivot
```

```python
erp_order_df=erp_order_str_pivot.merge(erp_order_num_pivot,how='left',on=['id','anonymous_id','event_name','at'])
erp_order_df.columns=erp_order_df.columns.str.replace('properties.',"")
erp_order_df=erp_order_df[['anonymous_id','at','order','total_value','discount_value']]
erp_order_df
```

### Lần mua hàng gần nhất

```python

erp_order_df['at']=pd.to_datetime(erp_order_df['at'])
last_order=erp_order_df.groupby('anonymous_id').agg({'at': np.max}).reset_index().sort_values('anonymous_id')
```

### Tổng số tiền mua hàng

```python
sales=erp_order_df.groupby('anonymous_id').agg({'total_value': sum}).sort_values('anonymous_id').reset_index()
sales=sales.rename(columns={'total_value':'M'})
```

### Tần suất mua hàng

```python
count_order=erp_order_df.groupby('anonymous_id').order.count().reset_index().sort_values('anonymous_id')
count_order=count_order.rename(columns={'order':'F'})
```

### Discount theo đơn hàng

```python
discount_order=erp_order_df.groupby('anonymous_id').discount_value.sum().reset_index().sort_values('anonymous_id')
discount_order
```

### Discount theo item sản phẩm


```python
#pivot dữ liệu numberic của nhóm item_order
erp_item_num=data_num_df[data_num_df['event_name'].isin(['erp_item_order_completed'])]
# 
# Tính tổng chiết khấu theo sản phẩm
erp_item_ck=erp_item_num[(erp_item_num['keys']=='properties.t_tien_sau_ck')|(erp_item_num['keys'].str.contains('gia_ban'))|(erp_item_num['keys'].str.contains('so_luong'))]
erp_item_num_pivot=erp_item_ck.pivot(index=['id','anonymous_id','event_name','at'],columns='keys',values='vals').reset_index()

# Tính tổng chiết khấu: số lượng x giá bán - tổng tiền sau chiết khấu
erp_item_num_pivot['tong_ck']=(erp_item_num_pivot['properties.gia_ban']*erp_item_num_pivot['properties.so_luong']-erp_item_num_pivot['properties.t_tien_sau_ck'])
erp_item_num_pivot_=erp_item_num_pivot[['anonymous_id','tong_ck']]
item_ck=erp_item_num_pivot.groupby('anonymous_id').tong_ck.sum().reset_index().sort_values('anonymous_id')
```

```python
item_ck=item_ck.rename(columns={'tong_ck':'D'})
item_ck
```
